## Data Cleaning


### Olympic athletes

```python
# Aggregate the data into one Data Frame 
import numpy as np
import pandas as pd

olympic_athletes = pd.read_csv('olympic_athletes.csv')
olympic_hosts = pd.read_csv('olympic_hosts.csv')
olympic_medals = pd.read_csv('olympic_medals.csv')
```

```python
olympic_athletes_df = pd.DataFrame(olympic_athletes)
olympic_hosts_df = pd.DataFrame(olympic_hosts)
olympic_medals_df = pd.DataFrame(olympic_medals)
```

```python
olympic_athletes_df.info()
```

```python
olympic_athletes_df = olympic_athletes_df.drop(columns=['bio', 'athlete_medals'])
```

```python
olympic_athletes_df.isna().sum()
```

```python
olympic_athletes_df.dropna(inplace=True)
```

```python
olympic_athletes_df
```

```python
olympic_athletes_df['athlete_full_name'] = olympic_athletes_df['athlete_full_name'].apply(lambda x:x.lower())
olympic_athletes_df['athlete_full_name'] = olympic_athletes_df.athlete_full_name.str.title()
```

```python
olympic_athletes_df=olympic_athletes_df.astype({'athlete_year_birth': 'int32'})
```

```python
olympic_athletes_df
```

```python
olympic_athletes_df[olympic_athletes_df.athlete_url.duplicated(keep=False)]
```

```python
olympic_athletes_df = olympic_athletes_df.drop_duplicates(subset=["athlete_url"], keep=False)
olympic_athletes_df
```

```python
olympic_athletes_df['first_game'] = olympic_athletes_df['first_game'].str.replace(' ','-')
```

```python
olympic_athletes_df
```

```python
olympic_athletes_df
```

```python
olympic_athletes_df.isna().sum()
```

```python
olympic_hosts_df.info()
```

```python
olympic_hosts_df.isna().sum()
```

```python
olympic_hosts_df['game_name'] = olympic_hosts_df['game_name'].str.replace(' ','-')
```

```python
olympic_hosts_df = olympic_hosts_df.drop(columns=['game_slug', 'game_end_date', 'game_start_date'])
```

```python
olympic_hosts_df['game_name'] = olympic_hosts_df['game_name'].str.replace(' ','-')
```

```python
olympic_hosts_df
```

```python
olympic_medals_df.info()
```

```python
olympic_medals_df = olympic_medals_df.drop(columns=['country_code', 'country_3_letter_code', 'participant_title', 'athlete_url'])
```

```python
olympic_medals_df
```

```python
olympic_medals_df.isna().sum()
```

```python
olympic_medals_df.dropna(inplace=True)
```

```python
olympic_medals_df.isna().sum()
```

```python
olympic_medals_df['slug_game'] = olympic_medals_df.slug_game.str.capitalize()
```

```python
olympic_medals_df['medal_type'] = olympic_medals_df['medal_type'].apply(lambda x:x.lower())
olympic_medals_df['medal_type'] = olympic_medals_df.medal_type.str.capitalize()
olympic_medals_df['participant_type'] = olympic_medals_df['participant_type'].astype(str).str.replace('GameTeam', 'Game Team')
```

```python
olympic_medals_df['event_title'] = olympic_medals_df.event_title.str.capitalize()
olympic_medals_df['athlete_full_name'] = olympic_medals_df.athlete_full_name.str.title()
```

```python
olympic_medals_df
```

```python
olympic_athletes_df.to_csv('olympic_athletes_df.csv')
olympic_hosts_df.to_csv('olympic_hosts_df.csv')
olympic_medals_df.to_csv('olympic_medals_df.csv')
```
